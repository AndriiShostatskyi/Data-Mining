## Programming Assignment # 1

This code visualizes data from “Table Data: Global and Hemispheric Monthly Means and Zonal Annual Means.”.
Namely, it plots the line chart where each line represents temperatures (Farenheit) over given years.

```python
import csv
import matplotlib.pyplot as plt
```

```python
FILE_NAME = 'ExcelFormattedGISTEMPDataCSV.csv'
```

### 1. Read data

```python
def read_csv_data(file_name):
    data = []
    with open(file_name) as csvfile:
        for row in csv.reader(csvfile, delimiter=' ', quotechar='|'):
            data.append(row[0].split(','))
    return data

data = read_csv_data(FILE_NAME)
```

### 2. Process data

```python
years = []
months_names = ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec')
months_data = [[] for _ in range(12)]

for row in data[1:-1]: 
    years.append(int(row[0]))
    for m in range(12):
        months_data[m].append(int(row[m+1]))
```

### 3. Plot data

```python
plt.figure(figsize=(16,12))
for i, n in enumerate(months_names):
    plt.plot(years, months_data[i], label=n)
plt.legend()
plt.xlabel("Year")
plt.ylabel("Temperature Farenheit"); 
plt.show()
```
